```python
from ultralytics import YOLO# Load the pre-trained YOLOv8 face detector from the YOLOv8-Face repository
pre_trained_model = YOLO('yolov8n.pt')  # Path to the pre-trained model from the GitHub repository
```

```python
import time
import os
import cv2

val_image_folder = '/home/cse/Documents/MdSourav/DNN/Assignment7/WIDER Face Dataset/split_data/val/images'  # Provide the correct path

# Function to run inference and calculate FPS
def run_inference(model, image_folder):
    image_files = os.listdir(image_folder)

    total_time = 0
    total_frames = 0

    for image_name in image_files:
        image_path = os.path.join(image_folder, image_name)
        img = cv2.imread(image_path)

        # Start the timer
        start_time = time.time()

        # Run inference on the image
        results = model(image_path)

        # Since results is a list, access the first result
        result = results[0]

        # End the timer
        end_time = time.time()
        total_time += (end_time - start_time)

        # Count the number of frames
        total_frames += 1

    # Calculate FPS
    fps = total_frames / total_time
    return total_time, fps

# Measure inference time and FPS for fine-tuned YOLOv8
fine_tuned_time, fine_tuned_fps = run_inference(fine_tuned_model, val_image_folder)

# Measure inference time and FPS for pre-trained YOLOv8n
pre_trained_time, pre_trained_fps = run_inference(pre_trained_model, val_image_folder)

print(f"Fine-Tuned YOLOv8 - Total Time: {fine_tuned_time:.2f} seconds, FPS: {fine_tuned_fps:.2f}")
print(f"Pre-trained YOLOv8n - Total Time: {pre_trained_time:.2f} seconds, FPS: {pre_trained_fps:.2f}")
```
